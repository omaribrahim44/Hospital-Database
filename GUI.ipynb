```python
import tkinter as tk
from tkinter import ttk
import mysql.connector
from tkinter import messagebox
from datetime import datetime
```

```python
db = mysql.connector.connect(
    host="localhost",
    user="root",
    passwd="yk182004#",
    database="project"
)

cursor = db.cursor()

def display_data(table_name):
    query = f"SELECT * FROM {table_name}"
    cursor.execute(query)
    data = cursor.fetchall()

    for row in tree_view.get_children():
        tree_view.delete(row)

    for row in data:
        tree_view.insert("", tk.END, values=row)

def insert_Doctor_data(Doctor_ID,First_name, Last_name, Zip_number, Street,City,Salary,Gender):
    query = "INSERT INTO Doctor (Doctor_ID,First_name, Last_name, Zip_number, Street,City,Salary,Gender) VALUES (%s,%s,%s,%s,%s,%s,%s,%s)"
    values = (Doctor_ID,First_name, Last_name, Zip_number, Street,City,Salary,Gender)
    cursor.execute(query, values)
    db.commit()
    messagebox.showinfo("Success", "Doctor data inserted successfully!")

root = tk.Tk()
root.title("Hospital Management System")


doctors_btn = tk.Button(root, text="Show Doctors", command=lambda: display_data("Doctor"))
doctors_btn.pack()

Medicine_btn = tk.Button(root, text="Show Medicine", command=lambda: display_data("Medicine"))
Medicine_btn.pack()

Equipment_btn = tk.Button(root, text="Show Equipment", command=lambda: display_data("Equipment"))
Equipment_btn.pack()

Receptionist_btn = tk.Button(root, text="Show Receptionist", command=lambda: display_data("Receptionist"))
Receptionist_btn.pack()

Room_btn = tk.Button(root, text="Show Room", command=lambda: display_data("Room"))
Room_btn.pack()

Nurse_btn = tk.Button(root, text="Show nurse", command=lambda: display_data("nurse"))
Nurse_btn.pack()

Clinics_btn = tk.Button(root, text="Show Clinics", command=lambda: display_data("Clinics"))
Clinics_btn.pack()

patient_btn = tk.Button(root, text="Show Patients", command=lambda: display_data("Patient"))
patient_btn.pack()

Doctor_ID_entry = tk.Entry(root)
Doctor_ID_entry.pack()

first_name_entry = tk.Entry(root)
first_name_entry.pack()

last_name_entry = tk.Entry(root)
last_name_entry.pack()

Zip_number_entry = tk.Entry(root)
Zip_number_entry.pack()

Street_entry = tk.Entry(root)
Street_entry.pack()

City_entry = tk.Entry(root)
City_entry.pack()

Salary_entry = tk.Entry(root)
Salary_entry.pack()

Gender_entry = tk.Entry(root)
Gender_entry.pack()

insert_btn = tk.Button(root, text="Insert Doctor Data", command=lambda: insert_Doctor_data(
    Doctor_ID_entry.get(),
    first_name_entry.get(),
    last_name_entry.get(),
    Zip_number_entry.get(),
    Street_entry.get(),
    City_entry.get(),
    Salary_entry.get(),
    Gender_entry.get(),
))
insert_btn.pack()


columns = ("", "", "", "","","","","","")
tree_view = ttk.Treeview(root, columns=columns, show="headings")

for col in columns:
    tree_view.heading(col, text=col)
    tree_view.column(col, width=120)

tree_view.pack()

root.mainloop()
```
